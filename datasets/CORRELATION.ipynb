```r
library('ggplot2')
library('datasets')
```

```r
ncbirths<-read.csv("ncbirths.csv")
```

```r
str(ncbirths)
```

```r
ggplot(data=ncbirths,aes(x=ncbirths$Weight..pounds.,y=Weeks.Gestation))+geom_point()
```

```r
ggplot(data=ncbirths,aes(x=cut(ncbirths$Weight..pounds.,breaks=5),y=Weeks.Gestation))+geom_point()
```

```r
ggplot(data=ncbirths,aes(x=cut(ncbirths$Weight..pounds.,breaks=5),y=Weeks.Gestation))+geom_boxplot()
```
